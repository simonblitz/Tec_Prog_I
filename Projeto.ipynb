2.Quais e quantos sÃ£o os cursos que abordam o assunto: JavaScript?

```python
import re
t = dt.fillna({"course_title": "NaN"})
js = t[t["course_title"].str.contains("JavaScript",flags=re.IGNORECASE, regex=True)]
js.loc[:,"course_title"]
```
