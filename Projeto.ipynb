```python
import pandas as pd
import numpy as np
```

```python
dt_bus = pd.read_csv(r"C:\Users\SIMONE\Desktop\LetsCode\Tec Prog\base de dados\3.1-data-sheet-udemy-courses-business-courses.csv")
dt_des = pd.read_csv(r"C:\Users\SIMONE\Desktop\LetsCode\Tec Prog\base de dados\3.1-data-sheet-udemy-courses-design-courses.csv")
dt_mus = pd.read_csv(r"C:\Users\SIMONE\Desktop\LetsCode\Tec Prog\base de dados\3.1-data-sheet-udemy-courses-music-courses.csv")
dt_web = pd.read_csv(r"C:\Users\SIMONE\Desktop\LetsCode\Tec Prog\base de dados\3.1-data-sheet-udemy-courses-web-development.csv")
```

```python
dt = pd.concat([dt_bus,dt_des,dt_mus,dt_web])
dt
```

1. Quantos cursos estão disponíveis na plataforma?

```python
q = dt.groupby(["course_title"]).count().reset_index()[["course_id"]].shape[0]
print(f"Tem {q} cursos disponiveis na plataforma.")
```

2.Quais e quantos são os cursos que abordam o assunto: JavaScript?

```python
import re
t = dt.fillna({"course_title": "NaN"})
js = t[t["course_title"].str.contains("JavaScript",flags=re.IGNORECASE, regex=True)]
js.loc[:,"course_title"]
```

```python
qjs = js["course_title"].shape[0]
print(f"{qjs} cursos abordam assunto JavaScript.")
```

3.Qual é o preço médio (coluna price) dos cursos oferecidos na plataforma?

```python
pr = np.round(dt['price'].mean(),2)
print(f"O preço médio é: {pr}")
```

4. Quais são os cinco cursos com maior número de inscritos (coluna num_subscribers)?

```python
p = dt.nlargest(5,["num_subscribers"])
p
```

5. Dos cinco cursos mais populares em termos de inscritos, mostre o rate médio, máximo e mínimo de cada um deles.

```python
rt = np.round(p["Rating"].mean(),2)
rmax = p["Rating"].max()
rmin = p["Rating"].min()

print(f"O rate médio é: {rt}, o rate máximo é: {rmax} e o rate mínimo é: {rmin}")
```
